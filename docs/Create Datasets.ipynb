```python
import pandas as pd
```

```python
data = pd.read_excel('datasets-new.xlsx');
data
```

```python
def cleaning(doc):
    doc = [tem.replace(u"\\xa0",u"") for tem in doc]
    doc = [tem.replace(u"\\r",u"") for tem in doc]
    doc = [tem.replace(u"\\n",u", ") for tem in doc]
    doc = [tem.replace(u"\r",u"") for tem in doc]
    doc = [tem.replace(u"\n",u", ") for tem in doc]
    doc = [tem.replace(u" ,",u"") for tem in doc]
    doc = [tem.replace(u".,",u".") for tem in doc]
    doc = [str(tem) for tem in doc]
    return doc
```

```python
nama = cleaning(data.Nama.tolist())
bahan = cleaning(data.Bahan.tolist())
resep = cleaning(data.Resep.tolist())
usia = cleaning(data.Usia.tolist())
```

```python
df = pd.DataFrame()
df
```

```python
df['Nama'] = nama
df['Bahan'] = bahan
df['Resep'] = resep
df['Usia'] = usia
df['Bahan'][112]
```

```python
df.to_excel('dataset-clean.xlsx', index=False)
```
